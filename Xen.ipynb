```python
#Let's Get the year,time, day, month, hour,minute, and second so that we can use as new features by creating a function

def extract_date(df,column):
    df[column+"_year"] = df[column].apply(lambda x:x.year)
    df[column+"_month"] = df[column].apply(lambda x:x.month)
    df[column+"_day"] = df[column].apply(lambda x:x.day)
    df[column+"_hour"] = df[column].apply(lambda x:x.hour)
    df[column+"_minute"] = df[column].apply(lambda x:x.minute)
    df[column+"_second"] = df[column].apply(lambda x:x.second)
```

## Data Preparation

Models such as linear and logistic regression need to have dummy variables for categorical variables, since those models (being mathematical equations) can process only numeric variables.

All that is not required for the Random Forest Model, since they can process categorical variables easily. However, we still need to encode the categorical variables into a standard format so that sklearn can understand them and build the tree. We'll do that using the LabelEncoder() class, which comes with sklearn.preprocessing.

```python
train.columns
```

```python
catcoltrain = train[['ChannelId','ProductCategory','ProviderId','ProductId','BatchId']]
```

```python
catcoltrain.head()
```

```python
catcoltest = test[['ChannelId','ProductCategory','ProviderId','ProductId','BatchId']]
```

```python
catcoltest.head()
```

```python
#We now Import the Necessary Library for Label Encoding
from sklearn import preprocessing
```

```python
#We then appply Label Encoding to the test and train Data Frames

le = preprocessing.LabelEncoder()
dftrain_categorical = catcoltrain.apply(le.fit_transform)
dftrain_categorical.head()
```

```python
dftest_categorical = catcoltest.apply(le.fit_transform)
dftest_categorical.head()
```

```python
train.head()
```

```python
test.head()
```

```python
#Dropping Encoded Features from the Original Data Frame
ft_drop = ['TransactionId', 'BatchId', 'AccountId', 'SubscriptionId', 'ProviderId','ProductId', 'ProductCategory', 'ChannelId']
```

```python
train.drop(ft_drop,axis=1,inplace=True)
test.drop(ft_drop,axis=1,inplace=True)
```

```python
train.head()
```

```python
test.head()
```

```python
# concat df_categorical with original df

train = pd.concat([train, dftrain_categorical], axis=1)
train.head()
```

## Model Building and Evaluation

Since at the beginning we saw that our FRAUD CASES RESULT is highly imbalanced we will have to over sample the Fraudelent Cases To Ensure our Model Performs Well

```python
#We will First import SMOTE
from imblearn.over_sampling import SMOTE
```

```python
# Separate input features and target

# Putting feature variable to X
X = train.drop(['FraudResult','TransactionStartTime'],axis=1)

# Putting response variable to y
y = train['FraudResult']
```

```python
X.columns
```

```python
# Importing train-test-split 
from sklearn.model_selection import train_test_split
```

```python
# setting up testing and training sets
X_train, X_vtest, y_train, y_vtest = train_test_split(X, y, test_size=0.25, random_state=27)
```

```python
sm = SMOTE(random_state=27, ratio=1.0)
X_train, y_train = sm.fit_sample(X_train, y_train)
```

```python
# Importing random forest classifier from sklearn library
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=500,criterion='entropy',max_features='auto',min_samples_leaf=2,min_samples_split=2)
```

```python
# train the Model
rfc.fit(X_train,y_train)
```

```python
# predict
y_pred = rfc.predict(X_vtest)
```

```python
# evaluation metrics
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_vtest,y_pred))
```

```python
print(confusion_matrix(y_vtest,y_pred))
```

```python
cm = confusion_matrix(y_vtest,y_pred)
```

```python
fig = plt.figure(figsize=(10,5))
sns.heatmap(cm,annot=True,fmt='d',cmap='YlGnBu')
```
